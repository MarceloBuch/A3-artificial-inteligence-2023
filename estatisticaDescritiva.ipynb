#### Os objetivos principais desta análise descritiva são:

2-	Avaliar Indicadores de Saúde: Explorar estatísticas relacionadas à saúde, como pressão arterial, colesterol, frequência cardíaca e índice de massa corporal (IMC), para identificar possíveis padrões ou anomalias.

3-	Investigar Hábitos de Vida e Comportamentais: Analisar variáveis relacionadas a hábitos de vida, como tabagismo, consumo de álcool, exercício físico e padrões de sono.


### Analisando os dados

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("./data/heart_attack_prediction_dataset.csv", sep=',')
df['systolic'] = df['Blood Pressure'].str.split('/').str[0]
df['diastolic'] = df['Blood Pressure'].str.split('/').str[1]
df = df.drop(columns=['Patient ID', 'Blood Pressure', 'Country', 'Continent', 'Hemisphere'])
df
```

```python
df.describe()
```

```python
df.info()
```

```python
paper = plt.figure(figsize=[20,10])
sns.heatmap(df.corr(),cmap='crest',annot=True)
plt.xlabel("Fatores de Risco")
plt.ylabel("Variáveis")
plt.title("Fatores de Risco")
plt.show()
```

Através dessa gráfico podemos analisar alguns pontos muito importantes e relevantes para a determinação do grau de risco de cada paciente, com base em diversos fatores de risco analisados. Quanto menor o número, menor o risco do paciente sofrer um ataque cardíaco e quando maior, mais probabilidade, é importante destacar que por mais que o risco 1 seja o nível mais alto de probabilidade, não necessariamente o paciente terá uma complicação cardíaca, essa analise de dados leva em consideração os dados analisados listados acima, porém existem fatores diferente nos quais não podemos prever
